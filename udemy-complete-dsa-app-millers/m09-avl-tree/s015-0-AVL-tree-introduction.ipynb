# What is an AVL Tree?

* An AVL tree is a **`self-balancing`** Binary Search Tree (BST) where the difference between heights of left and right subtrees cannot be more than one for all nodes.
* AVL tree is also a binary tree and all properties of binary tree will also be applicable to AVL tree.
* If at any time heights of left and right subtrees differ by more than one, then rebalancing is done to restore AVL property, this process is called **`ROTATION`** or **`HIEGHT-BALANCING`**.

# Why do we need AVL Tree?

Let say, we have numbers like: 10, 20, 30, 40, 50, 60, 70. 

**Binary Search Tree**

* Let insert the numbers in a binary search tree.
* In case of **Binary Search Tree**,
    * if the number is less than the root node, we insert the number as left subtree.
    * if the number is greater than the root node, we insert the number as right subtree.

After insertion, the Binary Search Tree will be a follows:



**Numbers**: 10, 20, 30, 40, 50, 60, 70
**Search for**: 60
**Time complexity is** `O(N)`

We know that, in case of Binary Search Tree, the time complexity for searching an element should be **`O(logN)`**. But, we got time complexity for searching an element as **`O(N)`**. **WHY?**

That is because the hieght of the tree is too much.

---

**AVL Tree (Balanced Binary Tree)**

If we balance the binary search tree, then it becomes an **AVL tree** and will be a follows:



**Numbers**: 10, 20, 30, 40, 50, 60, 70
**Search for**: 60
**Time complexity is** `O(logN)`

---


**Conclusion:** 
* AVL tree (**Balanced Binary Search Tree**) gives better performance while insertion/deletion/searching as compared to **Unbalanced Binary Search Tree**.
* The main reason for performance issue in **Unbalanced Binary Search Tree** is the increased height which degrades its performance.

**How AVL tree resolves the performance issue?**
* AVL tree ensure that the **Binary Search Tree** never gets into an **Unbalanced State**.
* AVL tree enforces rule called **ROTATION** to ensure that the  **Binary Search Tree** should always remain **Balanced** during insertion/deletion.
